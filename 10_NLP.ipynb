```python
import re
import random
```

```python
patterns = {
    r'hi|hello|hey': ["Hello ","Hi there!","Hey! how can I help you?"],
    r'how are you': ["I'm fine!", "Doing great!", "I am here to help you"],
    r'your name': ["I am a NLP chatbot", "You can call me ChatBot"],
    r'bye|exit' : ["Goodbye!", "See you later!", "Bye! Have a nice day"]
}
```

```python
def chatbot_response(user_input):
    user_input = user_input.lower()

    for key in patterns:
        # Use re.search for more flexible pattern matching
        if re.search(key, user_input):
            return random.choice(patterns[key])

    return "Sorry, I didn't understand that."
```

```python
while True:
    user = input("You: ")
    if "bye" in user.lower():
        print("Bot:", chatbot_response(user))
        break
    print("Bot:", chatbot_response(user))
```
